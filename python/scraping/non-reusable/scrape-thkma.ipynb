```python
# Process Page

def process_page(url):
    with open(outputPath, 'a+', encoding='utf8') as csvfile:
        r = requests.get(url)
        print(url)
        soup = BeautifulSoup(r.text, 'html.parser')

        name_div = soup.find('div', class_='name')
        if name_div is None:
            return

        name_en = name_div.find_all('p')[0].get_text(strip=True)
        name_cn = name_div.find_all('p')[1].get_text(strip=True)

        rows = soup.select('.info .row')
        data = {
            'id': url,
            'Name (English)': name_en,
            'Name (Chinese)': name_cn,
        }

        for row in rows:
            head = row.find('div', class_='head').get_text(strip=True)
            data_field = row.find('div', class_='data').get_text(strip=True)
            data[head] = data_field

        fieldnames = list(data.keys())
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writerow(data)
```

```python
import requests
from bs4 import BeautifulSoup
import csv

base_url = "https://www.thkma.org%s"
    
with open(outputPath, 'a+', encoding='utf8') as csvfile:
    spamwriter = csv.writer(csvfile, quoting=csv.QUOTE_MINIMAL)
    spamwriter.writerow(headers)

with open(linkOutputPath, 'r', encoding='utf8') as linkFile:
    for l in linkFile:
        url = base_url%(l.strip())
        process_page(url)
```
