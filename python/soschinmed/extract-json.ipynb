```python
import json
import requests
```

```python
res = {
    'itemsTotal': 0,
    'discountsTotal': 0,
    'total': 0,
    'hcv': 0,
}
for i in range(1, 31):
    d = str(i).zfill(2)
    s = requests.get('https://api.portal.soschinmed.clinic/report/daily/summary/2023/04/%s?clinic=1' % d)
    r = s.json()
    res['itemsTotal'] += r['itemsTotal']
    res['discountsTotal'] += r['discountsTotal']
    res['total'] += r['total']
    res['hcv'] += r['hcv']
print(res)
```
